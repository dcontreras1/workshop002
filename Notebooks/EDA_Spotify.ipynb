# Análisis Exploratorio de Datos Spotify

Este Notebook incluye estadísticas descriptivas, visualización de distribuciones, correlaciones y análisis de artistas y popularidad.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
df_spotify = pd.read_csv("/home/dcontreras/Workshop002/dataset.csv")
df_spotify.head()
```

Información general y valores nulos

```python
df_spotify.info()
df_spotify.isnull().sum()
```

Estadísticas descriptivas

```python
df_spotify.describe()
```

Artistas más frecuentes

```python
top_artists = df_spotify['artists'].value_counts().head(10)
top_artists.plot(kind='barh', figsize=(8, 5), color='coral')
plt.title("Top 10 Artistas más frecuentes")
plt.xlabel("Número de canciones")
plt.ylabel("Artista")
plt.show()
```

Distribución de la popularidad

```python
sns.histplot(df_spotify['popularity'], bins=20, kde=True, color='skyblue')
plt.title("Distribución de la popularidad")
plt.xlabel("Popularidad")
plt.ylabel("Frecuencia")
plt.show()
```

Matriz de correlación entre variable numéricas

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df_spotify.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Matriz de correlación Spotify")
plt.show()
```
