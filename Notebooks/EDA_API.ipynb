```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
file_path = '/home/dcontreras/Workshop002/temp/spotify_API.csv'
df_api = pd.read_csv(file_path)

df_api.head()
```

Información general

```python
df_api.info()
df_api.describe(include='all')
```

Revisión de valores nulos

```python
nulls = df_api.isnull().sum().sort_values(ascending=False)
nulls[nulls > 0]
```

Distribución de energía y valencia

```python
plt.figure(figsize=(10, 6))

# Distribución de energy
sns.kdeplot(df_api['energy'], color='blue', label='Energía', shade=True)

# Distribución de valence
sns.kdeplot(df_api['valence'], color='orange', label='Valencia', shade=True)

plt.title('Distribución de Energía y Valencia de las canciones')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.tight_layout()
plt.show()
```

Top 10 Géneros

```python
# Agrupar por género y calcular promedio de popularidad
top_genres = df_api.groupby('track_genre')['popularity'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='coolwarm')
plt.title('🎵 Popularidad promedio por género (Top 10)')
plt.xlabel('Popularidad promedio')
plt.ylabel('Género musical')
plt.tight_layout()
plt.show()
```

Popularidad promedio por si la canción es explícita o no

```python
plt.figure(figsize=(6, 5))
sns.boxplot(data=df_api, x='explicit', y='popularity', palette='Set2')
plt.title('🚫 Popularidad según si la canción es explícita')
plt.xlabel('Contenido explícito')
plt.ylabel('Popularidad')
plt.tight_layout()
plt.show()
```
